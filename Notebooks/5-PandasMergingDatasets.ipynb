### Merging two datasets

```python
import pandas as pd
```

```python
df1= pd.DataFrame({'id': ['A', 'B', 'C', 'D'], 'var1': [37, 36, 43, 23], 'var2': [120, 117, 230, 315]})
```

```python
df1
```

```python
df2 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'], 'var3': [12, 16, 13, 18], 'var4': [74, 54, 21, 36]})
```

```python
df2
```

```python
df3 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'], 'var3': [2, 6, 3, 8], 'var4': [7, 5, 2, 6]})
```

```python
df3
```

### use append() to join

```python
df1.append(df2, sort=False)
```

```python
df3.append(df1, sort=False)
```

```python
df3.append(df1, sort=True)
```

### use concat([datasets...]) to merge more than one dataset

```python
pd.concat([df1, df2, df3], sort=True)
```

### we can concatenate horizontaly by changing axis to 1 (default is 0)

```python
pd.concat([df1, df2, df3], sort=True, axis=1)
```
